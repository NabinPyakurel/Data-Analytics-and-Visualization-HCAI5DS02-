3. Describing The Data - Descriptive Statistics A Review.
 2. Patient Blood Pressure Readings:

```python
#Context: A health researcher is analyzing systolic blood pressure levels from a clinic.
#Systolic Blood Pressure
#118 122 125 130 135 138 142 144 146 150 152 155 160 162 165

bp = np.array([118, 122, 125, 130, 135, 138, 142, 144, 146, 150, 152, 155, 160, 162, 165])
print("Descriptive Statistics for Systolic Blood Pressure:",bp)
```

```python
#Calculating Mean
bpMean=np.mean(bp)
print("Mean:",bpMean)

#Calculating Median
bpMedian=np.median(bp)
print("Median:",bpMedian)

#Calculating Standard Deviation
bpStd=np.std(bp)
print("Standard Deviation:",bpStd)
```

```python
#Plot a histogram. Is the distribution skewed?

print("\nAnalysis of Skewness:")
# Calculate skewness
from scipy.stats import skew
bpSkew = skew(bp)
print(f"Skewness: {bpSkew:.4f}")

if bpSkew > 0:
  print("The distribution appears to be positively skewed (right-skewed).\n")
elif bpSkew < 0:
  print("The distribution appears to be negatively skewed (left-skewed).\n")
else:
  print("The distribution appears to be symmetric.\n")

  plt.figure(figsize=(8, 6))
plt.hist(bp, bins=5, color='skyblue', edgecolor='black')
plt.title('Histogram of Systolic Blood Pressure')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Compute the IQR. Identify any patients with unusually high blood pressure.

# Calculating the IQR
q1 = np.percentile(bp, 25)
q3 = np.percentile(bp, 75)
iqr = q3 - q1

print(f"Q1: {q1}")
print(f"Q3: {q3}")
print(f"IQR: {iqr}")

# Identifying unusually high blood pressure using IQR
upper_bound = q3 + 1.5 * iqr
print(f"\nUpper bound for unusual blood pressure (Q3 + 1.5*IQR): {upper_bound}")

unusually_high_bp = bp[bp > upper_bound]

if len(unusually_high_bp) > 0:
    print("Patients with unusually high blood pressure:")
    print(unusually_high_bp)
else:
    print("No patients with unusually high blood pressure based on the IQR method.")
```

A normal distribution is characterized by symmetry (skewness close to 0),
with the mean, median, and mode being approximately equal, and most data points
 falling within a few standard deviations of the mean (Empirical Rule: ~68% within 1 std dev, ~95% within 2 std devs, ~99.7% within 3 std devs).

#Based on the calculated statistics:
 1. Skewness: The skewness ({bpSkew:.4f}) is close to 0 (or slightly positive),
    suggesting the distribution is relatively symmetric, which is a characteristic of a normal distribution.
 2. Mean and Median: The mean ({bpMean:.2f}) and median ({bpMedian:.2f})
    are very close, which also indicates symmetry, another feature of a normal distribution.
 3. Histogram: The histogram visually appears somewhat symmetric, although with a small number of bins
    it's harder to definitively say.
 4. IQR and Outliers: Using the IQR method, there are no data points identified as unusually high
    blood pressure. While the absence of outliers doesn't guarantee normality,
    the presence of significant outliers can suggest non-normality.

 Conclusion:
 While we have a relatively small sample size (n=15), the available descriptive statistics (skewness close to zero, mean and median being similar, and the shape of the histogram) suggest that the distribution of systolic blood pressure in this clinic population appears to be approximately normal. However, with a larger sample size, formal normality tests (like the Shapiro-Wilk test) would provide a more robust assessment.

# Identify which summary statistics might be misleading if data is bimodal in a line

Mean, median, and mode can be misleading if data is bimodal because they each represent only one peak or the overall center, failing to capture the existence and location of both distinct modes.

# Suggest how the five-number summary could help identify immune outliers in a line

The five-number summary (minimum, first quartile (Q1), median, third quartile (Q3), and maximum) helps identify immune outliers by providing key thresholds. The Interquartile Range (IQR = Q3 - Q1) is used to calculate fences: Lower Fence = Q1 - 1.5 * IQR and Upper Fence = Q3 + 1.5 * IQR. Data points falling below the Lower Fence or above the Upper Fence are considered potential outliers. For "immune outliers," which might represent unusually high or low immune responses (like antibody levels), checking values outside these calculated fences based on the majority of the data can pinpoint individuals with significantly different responses.

**Discuss: How would a flash sale or promotional campaign distort the mean? How can you account for it in visuals or summaries?**


A flash sale or promotional campaign would likely lead to a significant spike in sales data during the campaign period. This influx of unusually high sales values would act as outliers, pulling the mean sales figure upwards. The mean, being sensitive to extreme values, would become an inflated representation of the typical daily sales during a regular, non-promotional period. The median, on the other hand, is less affected by extreme values and would likely remain a better indicator of the typical sales on a standard day. The distortion arises because the mean would be heavily influenced by the brief, high-volume period of the sale, masking the underlying average sales outside of the promotion.

To account for this distortion in visuals and summaries, several strategies can be employed. In summaries, it is crucial to report the mean alongside the median. Comparing the mean and median can immediately signal the potential presence of outliers or skewness introduced by the sale. Calculating and reporting the mean *excluding* the promotional period would provide a clearer picture of typical sales. In visualizations, time series plots are particularly useful; the flash sale would appear as a prominent spike, clearly indicating an unusual event. Annotating this spike with the campaign details further clarifies the context. Boxplots could show the distribution of sales values, where the promotional days might appear as high outliers above the whiskers. Histograms could show a skewed distribution with a tail extending towards higher values due to the sale. Additionally, creating separate analyses or visualizations specifically for the promotional period versus non-promotional periods allows for a direct comparison and highlights the impact of the campaign without letting it distort the overall picture. Utilizing moving averages with appropriate window sizes can also help smooth out the impact of a short-term sale, revealing underlying trends, although a very short sale might still significantly impact the moving average during and immediately after the event.

The Working Professionals segment also exhibits higher spending tendencies, as its density curve is shifted towards higher transaction amounts, and its peak is further to the right than the Student segment's peak.

```python
#Compute mean, median, IQR, and CV for each group and display them in a summary table.

def iqr_from_series(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    return q3 - q1

def cv_from_series(series):
    mean = series.mean()
    std = series.std()
    if mean == 0:
        return np.nan # Avoid division by zero
    return (std / mean) * 100

summary_table = df2.groupby('segment')['transaction_amount'].agg(
    Mean='mean',
    Median='median',
    IQR=iqr_from_series,
    CV=cv_from_series
).reset_index()

print("\nSummary Table of Transaction Amount by Segment:")
print(summary_table.round(2))
```

```python
#Construct a boxplot with swarm overlay to visualize typical vs. exceptional spending behavior. Do the visuals agree with summary statistics?

plt.figure(figsize=(10, 6))
sns.boxplot(x='segment', y='transaction_amount', data=df2, palette='pastel')
sns.swarmplot(x='segment', y='transaction_amount', data=df2, color='black', alpha=0.5)
plt.title('Boxplot with Swarm Overlay of Transaction Amount by Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Transaction Amount ($)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Yes, the visuals (boxplot with swarm overlay) agree with the summary statistics. The boxplot visually confirms that the 'Working Professionals' segment has a higher median and mean transaction amount than the 'Student' segment, aligning with the numerical summary table. The longer upper whisker and the presence of numerous points (from the swarm plot) extending higher for 'Working Professionals' compared to 'Students' visually represent the greater spread and potential for higher transaction amounts in that segment, which is supported by its higher calculated standard deviation and IQR in the summary statistics. The swarm plot also shows the individual data points, allowing for a visual assessment of the density and presence of potential outliers, which further reinforces the insights gained from the summary statistics regarding the variability and typical spending of each group.

```python
#Create a scatter plot of number of transactions vs. total spending for each customer, colorcoded by segment. What does the relationship suggest about loyalty or value?

customer_spending = df2.groupby(['customer_id', 'segment']).agg(
    total_spending=('transaction_amount', 'sum'),
    num_transactions=('transaction_amount', 'count')
).reset_index()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_spending, x='num_transactions', y='total_spending', hue='segment', palette='pastel')
plt.title('Scatter Plot of Number of Transactions vs. Total Spending by Segment')
plt.xlabel('Number of Transactions')
plt.ylabel('Total Spending ($)')
plt.legend(title='Customer Segment')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

The scatter plot visualizing the relationship between the number of transactions and total spending for each customer, color-coded by segment, provides insights into customer loyalty and value. Generally, a positive correlation where higher numbers of transactions correspond to higher total spending would suggest that frequent customers are also valuable customers. Comparing the segments, if one segment shows a steeper upward trend (meaning each additional transaction contributes more to total spending) or if customers in that segment achieve higher total spending with a similar number of transactions compared to the other segment, it suggests that segment is either more valuable per transaction or contains more loyal customers who make repeat, higher-value purchases. Observing clusters of points with high transaction counts and high total spending for a particular segment would indicate a group of highly loyal and valuable customers within that segment.
